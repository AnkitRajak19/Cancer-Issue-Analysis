#1.Find Out Smoking Status.

```python
smoking_counts = df['SmokingStatus'].value_counts()
```

```python
fig = px.pie(
    values=smoking_counts.values,
    names=smoking_counts.index,
    title='Smoking Status Distribution',
    labels={'values':'Number of People', 'names':'Smoking Status'}
)
fig.update_traces(textinfo='percent+label')
fig.show()
```

#2.Find Out Count Of Race/Ethnicity And Hospital Region

```python
race_ethnicity_counts = df['Race/Ethnicity'].value_counts()
hospital_region_counts = df['HospitalRegion'].value_counts()
```

```python
fig_race_ethnicity = px.bar(
    x=race_ethnicity_counts.index,
    y=race_ethnicity_counts.values,
    title='Count of Race/Ethnicity',
    labels={'x': 'Race/Ethnicity', 'y': 'Count'}
)
fig_race_ethnicity.show()


```

```python
fig_hospital_region = px.bar(
    x=hospital_region_counts.index,
    y=hospital_region_counts.values,
    title='Count of HospitalRegion',
    labels={'x': 'HospitalRegion', 'y': 'Count'}
)
fig_hospital_region.show()
```

#3.Find Out Smoking Status Distribution By Gender

```python
smoker_counts = df.groupby(['Gender', 'SmokingStatus'])['SmokingStatus'].count().reset_index(name='count')

smoker_counts['Gender_SmokingStatus'] = smoker_counts['Gender'] + ' - ' + smoker_counts['SmokingStatus']


fig = px.pie(
    smoker_counts,
    values='count',
    names='Gender_SmokingStatus',  # Use the combined label for names
    title='Smoking Status Distribution by Gender',
    labels={'count':'Number of People', 'Gender_SmokingStatus':'Gender and Smoking Status'},
    hole=0.3
)
fig.update_traces(textinfo='percent+label')
fig.show()
```

#4.Find OutDistribution Of Cancer Types.

```python
cancer_gender_counts = df.groupby(['CancerType', 'Gender'])['CancerType'].count().reset_index(name='Count')

fig_cancer_type = px.bar(
    cancer_gender_counts,
    x='CancerType',
    y='Count',
    color='Gender',  # Differentiate by gender using color
    barmode='group',  # Group bars by cancer type
    title='Distribution of Cancer Types by Gender',
    labels={'x': 'Cancer Type', 'y': 'Count', 'Gender': 'Gender'}
)

fig_cancer_type.show()
```

#5.Find OutDistribution Of Cancer Stage.

```python
stage_gender_counts = df.groupby(['Stage', 'Gender'])['Stage'].count().reset_index(name='Count')

fig_stage = px.bar(
    stage_gender_counts,
    x='Stage',
    y='Count',
    color='Gender',  # Differentiate by gender using color
    barmode='group',  # Group bars by stage
    title='Distribution of Cancer Stages by Gender',
    labels={'Stage': 'Stage', 'Count': 'Count', 'Gender': 'Gender'}
)

fig_stage.show()
```

#6.Find Out Distribution Of Tumor Size.

```python
tumor_size_gender_counts = df.groupby(['Gender', 'TumorSize'])['TumorSize'].count().reset_index(name='Count')

fig = px.bar(
    tumor_size_gender_counts,
    x='TumorSize',
    y='Count',
    color='Gender',
    barmode='group',  # Group bars by gender
    title='Distribution of Tumor Sizes by Gender',
    labels={'TumorSize': 'Tumor Size', 'Count': 'Count', 'Gender': 'Gender'}
)

fig.show()
```

#7.Find Out Distribution Of Cancer Types By Gender.

```python
cancer_gender_counts = df.groupby(['Gender', 'CancerType'])['CancerType'].count().reset_index(name='Count')
```

```python
fig = px.line(
    cancer_gender_counts,
    x='CancerType',
    y='Count',
    color='Gender',
    title='Distribution of Cancer Types by Gender'
)

fig.update_layout(
    xaxis_title='Cancer Type',
    yaxis_title='Count',
    xaxis={'categoryorder':'total descending'}  # Sort x-axis by total count
)

fig.show()
```

#8.Find Out Treatment Type Distribution By Gender.

```python
treatment_gender_counts = df.groupby(['Gender', 'TreatmentType'])['TreatmentType'].count().reset_index(name='Count')
```

```python
fig = px.scatter(
    treatment_gender_counts,
    x="TreatmentType",
    y="Count",
    color="Gender",
    title="Treatment Type Distribution by Gender"
)

fig.show()
```

#9.Find Out Recurrence And Survival Month By Gender.

```python
recurrence_survival_counts = df.groupby(['Gender', 'Recurrence'])['SurvivalMonths'].count().reset_index(name='Count')
```

```python
fig = px.bar(
    recurrence_survival_counts,
    x='Recurrence',
    y='Count',
    color='Gender',
    barmode='group',
    title='Recurrence and Survival Month by Gender',
    labels={'Recurrence': 'Recurrence', 'Count': 'Count', 'Gender': 'Gender'}
)

fig.show()
```

#10.Find Out Distribution Of Survival Months By Gender.

```python
fig = px.histogram(
    df,
    x="SurvivalMonths",
    color="Gender",
    barmode="group",  # To group bars by gender
    title="Distribution of Survival Months by Gender",
    labels={"SurvivalMonths": "Survival Months", "Gender": "Gender"},
    nbins=20  # Adjust the number of bins as needed
)
fig.show()
```
